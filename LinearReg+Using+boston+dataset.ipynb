```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn 
from sklearn.datasets import load_boston
```

```python
boston=load_boston()
boston_df=pd.DataFrame(boston.data)
df2=pd.DataFrame(boston.target)
boston_df
```

lets build a linear regression model uding all 13 attributes x(independnet variables) that will predict dependent(Y)

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(boston_df.iloc[:,:13],boston_df.iloc[:,-1])
```

```python
print(lr.intercept_)
```

```python
print(lr.coef_)
```

```python
coef_series=pd.Series(lr.coef_,index= boston.feature_names)
print(coef_series)
```

```python
plt.scatter(boston_df.RM,boston_df.actual_price)
plt.xlabel("RM")
plt.ylabel("actual price")
plt.title("RM vs actual price")
plt.show()
```

```python
plt.scatter(boston_df.NOX,boston_df.actual_price)
plt.xlabel("NOX")
plt.ylabel("actual price")
plt.title("NOX vs actual price")
plt.show()
```

```python
predictions=lr.predict(boston_df.iloc[:,:13])
```

```python
plt.scatter(predictions,boston_df.actual_price)
plt.xlabel="predictions"
plt.ylabel="Actual prices"
plt.show()
```

```python
Errors=pd.Series(boston_df.actual_price-predictions)
print(np.mean(Errors**2))
```

```python
X_train,X_test,Y_train,Y_test=sklearn.model_selection.train_test_split(boston_df.iloc[:,:13],boston.target,test_size=0.25,random_state=5)
```

```python
lr1=LinearRegression()
lr1.fit(X_train,Y_train)
pred_train=lr1.predict(X_train)
pred_test=lr1.predict(X_test)
```

```python
print(np.mean((Y_train-pred_train)**2))
```

```python
(print(np.mean((Y_test-pred_test)**2)))
```

```python
plt.scatter(pred_train,pred_train-Y_train,c='b',s=40,alpha=0.5)
plt.scatter(pred_test,pred_test-Y_test,c='g',s=40)
plt.hlines(y=0,xmin=-10,xmax=50)
plt.vlines(x=0,ymin=-20,ymax=50)
plt.show()
```
